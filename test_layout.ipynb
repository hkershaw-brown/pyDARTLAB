```python
import seaborn as sns; sns.set_theme()
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
sns.rugplot(data=tips, x="total_bill", y="tip", hue="time")
```

```python
import plotly.express as px
df = px.data.tips()
#fig = px.ecdf(df, x="total_bill", color="sex", markers=True, lines=False, marginal="rug")
fig = px.scatter(df, x="total_bill",y="tip", marginal_y="rug", marginal_x="rug", color="sex")


fig.show()
```
